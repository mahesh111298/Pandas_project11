```python
#2. What was the average age of the passengers on the Titanic?
# Calculate mean age of passengers
avg_age = np.mean(titanic_df['age'])
print("Average age of passengers: ", avg_age)
```

```python
#5. What was the average fare paid by the passengers on the Titanic?
# Calculate average fare
avg_fare = np.mean(titanic_df['fare'])
print("Average fare paid by passengers: ", avg_fare)
```

```python
#6. What was the survival rate of male passengers on the Titanic?
# Subset male passengers
male_passengers = titanic_df[titanic_df['sex'] == 'male']

# Calculate survival rate
male_survival_rate = np.mean(male_passengers['survived'])
print("Survival rate of male passengers: ", male_survival_rate)
```

```python
#7. What was the survival rate of female passengers on the Titanic?
# Subset female passengers
female_passengers = titanic_df[titanic_df['sex'] == 'female']

# Calculate survival rate
female_survival_rate = np.mean(female_passengers['survived'])
print("Survival rate of female passengers: ", female_survival_rate)

 
```

```python
#8. What was the median age of male passengers on the Titanic?
# Subset male passengers
male_passengers = titanic_df[titanic_df['sex'] == 'male']

# Calculate median age
male_median_age = np.median(male_passengers['age'])
print("Median age of male passengers: ", male_median_age)
```

```python
#9. What was the median age of female passengers on the Titanic?
# Subset female passengers
female_passengers = titanic_df[titanic_df['sex'] == 'female']

# Calculate median age
female_median_age = np.median(female_passengers['age'])
print("Median age of female passengers: ", female_median_age)
```

```python
# 20  different graphs
#1.Bar Chart - Number of Passengers by Gender?
import seaborn as sns

sns.countplot(x='sex', data=titanic_df)
```

```python
# 2.Histogram - Distribution of Age of Passengers?
sns.histplot(x='age', data=titanic_df, bins=30)
```

```python
# 3.Box Plot - Age by Passenger Class?
sns.boxplot(x='pclass', y='age', data=titanic_df)
```

```python
#4.Violin Plot - Age by Passenger Class?
sns.violinplot(x='pclass', y='age', data=titanic_df)
```

```python
# 5.Scatter Plot - Age vs Fare ?
sns.scatterplot(x='age', y='fare', data=titanic_df)
```

```python
#6.Heat Map - Correlation between Features?
sns.heatmap(titanic_df.corr(), annot=True)
```

```python
#7.Pie Chart - Proportion of Passengers by Gender?
gender_counts = titanic_df['sex'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
```

```python
#8.Stacked Bar Chart - Number of Passengers by Gender and Survival?
sns.countplot(x='survived', hue='sex', data=titanic_df)
```

```python
#10.Pair Plot - Pairwise Relationships between Features?
sns.pairplot(data=titanic_df, vars=['age', 'fare', 'pclass'], hue='survived')
```

```python
#11.Line Chart - Survival Rate by Passenger Class?
survival_rates = titanic_df.groupby('pclass')['survived'].mean()
sns.lineplot(x=survival_rates.index, y=survival_rates)
```

```python
#18 Pair Grid - Pairwise Relationships between Features with Regression Line?
g = sns.PairGrid(data=titanic_df, vars=['age', 'fare', 'pclass'])
g.map_upper(sns.scatterplot)
g.map_lower(sns.regplot)
g.map_diag(sns.histplot)

```
